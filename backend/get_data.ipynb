```python
import numpy as np
import matplotlib.pylab as plt
from PIL import Image

path = "./data/"
WIDTH = 512
HEIGHT = 32
img_size = (WIDTH, HEIGHT)
def get_image(path):
    img = Image.open(path).convert('RGB')
    img = img.crop((0,0,img.width,min(img.height,int(0.06*img.width))))
    img = img.resize(img_size)
    img = np.array(img)
    return img
```

```python
name=path+"oneui/"+"oneui_75.jpg"
img = Image.open(name).convert('RGB')

plt.imshow(get_image(name))
plt.show()
plt.figure(figsize=(60,1))
#dep=52
#dep=27
dep=10
dep=25
img = img.crop((0,dep,img.width,dep+int(0.06*img.width)))
plt.imshow(np.array(img))
plt.show()
```
